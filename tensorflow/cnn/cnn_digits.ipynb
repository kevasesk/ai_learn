```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_sample_image

china = load_sample_image("china.jpg")
flower = load_sample_image("flower.jpg")

dataset = np.array([china, flower], dtype=np.float32)
batch_size, height, width, channels = dataset.shape

filter_side_size = 7
stride = 1

filters = np.zeros(shape=(filter_side_size, filter_side_size, channels, 2), dtype=np.float32)
filters[:, 3, :, 0] = 1
filters[3, :, :, 1] = 1 

conv_layer = tf.keras.layers.Conv2D(
    filters=2,
    kernel_size=(filter_side_size, filter_side_size),
    strides=(stride, stride),
    #padding="same", 
    use_bias=False,
    kernel_initializer=tf.constant_initializer(filters) 
)
output = conv_layer(dataset)

print(batch_size, height, width, channels)
print(output.shape) 

plt.imshow(output[0, ..., 0], cmap="gray")

```
