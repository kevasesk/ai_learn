```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()

# Extract features and target
sepal_length = iris.data[:, 0]  # Feature 0
sepal_width = iris.data[:, 1]   # Feature 1
petal_length = iris.data[:, 2]  # Feature 2
petal_width = iris.data[:, 3]   # Feature 3
species = iris.target           # Species labels (0, 1, 2)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Define the pairs to plot (data, labels)
pairs = [
    (sepal_length, sepal_width, 'Sepal Length (cm)', 'Sepal Width (cm)'),
    (sepal_length, petal_length, 'Sepal Length (cm)', 'Petal Length (cm)'),
    (sepal_width, petal_length, 'Sepal Width (cm)', 'Petal Length (cm)'),
    (petal_length, petal_width, 'Petal Length (cm)', 'Petal Width (cm)')
]

# Plot each pair in the corresponding subplot
for ax, (x, y, xlabel, ylabel) in zip(axs.flat, pairs):
    # Plot points for each species separately to assign colors and labels
    for sp in np.unique(species):
        mask = species == sp
        ax.scatter(x[mask], y[mask], label=iris.target_names[sp])
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
```
